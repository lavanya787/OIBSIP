```python
# Aggregate sales by date
sales_by_date = df.groupby('Date')['Price per Unit'].sum().reset_index()

# Plotting sales over time
plt.figure(figsize=(12, 6))
plt.plot(sales_by_date['Date'], sales_by_date['Price per Unit'], marker='o')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Price Per Unit')
plt.show()

# Optional: Decompose the time series if needed
# from statsmodels.tsa.seasonal import seasonal_decompose
# decomposition = seasonal_decompose(sales_by_date['Sales'], model='additive', period=12)
# decomposition.plot()
# plt.show()
```

```python
df=df.dropna()
```

```python
df.tail()
```

```python
# Example: Sales distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Product Category'], kde=True)
plt.title('Distribution of Sales')
plt.show()
```

```python
# Example: Analyzing sales by customer segments
customer_sales = df.groupby('Customer ID')['Price per Unit'].sum().reset_index()
print(customer_sales.describe())

# Example: Analyzing product sales
product_sales = df.groupby('Product Category')['Price per Unit'].sum().reset_index()
top_products = product_sales.sort_values(by='Price per Unit', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Product Category', y='Price per Unit', data=top_products)
plt.title('Top 10 Products by Sales')
plt.show()
```
