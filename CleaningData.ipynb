```python
df_clean=df.dropna()
df_clean
```

```python
df_clean['MathScore'].plot(kind='hist', bins=20, title='MathScore')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df_clean=df_clean.drop_duplicates()
df_clean
```

```python
df_clean.boxplot(column='MathScore', by=['Gender', 'TestPrep'])
plt.xlabel('Gender and Test Preparation')
plt.ylabel('Math Score')
_ = plt.title('Distribution of Math Scores by Gender and Test Preparation')
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.head()
```

```python
z_scores = np.abs(stats.zscore(df_clean['ReadingScore']))
outliers = df_clean[(z_scores > 3)]
outliers
```

```python
#Remove outliers
df_clean = df_clean[(z_scores <= 3)]
```

```python
df_clean['WritingScore'] = np.where(z_scores > 3, df_clean['WritingScore'].mean(), df_clean['WritingScore'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['ReadingScore'])
plt.title('Boxplot before removing outliers')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_clean['ReadingScore'])
plt.title('Boxplot after removing outliers')
plt.show()
```
